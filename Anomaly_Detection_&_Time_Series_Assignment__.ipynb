# **Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.**

# **Answer1.**
Anomaly Detection (also called outlier detection) is the process of identifying data points, patterns, or events that deviate significantly from the normal behavior of a dataset.
These anomalies often indicate errors, rare events, or important insights such as fraud, equipment failure, or unexpected behavior in systems.

Types of Anomalies
Point Anomalies

Definition: A single data point that is very different from the rest of the data.

When it occurs: Common in datasets where one observation does not follow the general trend.

Example:

In a bank transaction dataset, if a person usually spends ₹2,000–₹10,000 per month but suddenly spends ₹1,00,000, that transaction is a point anomaly.

In temperature data, if the usual range is 20°C–30°C and one reading shows 55°C, it’s a point anomaly.

Contextual Anomalies

Definition: A data point that may be normal in general, but considered anomalous in a specific context.

When it occurs: Usually in time-series or spatial data where context (time, location, or situation) matters.

Example:

A temperature of 30°C is normal in summer but abnormal in winter → contextual anomaly.

A sudden spike in website traffic at 3 AM could be suspicious, even though the same spike at 6 PM is normal.

3. Collective Anomalies

Definition: A group of related data points that together behave abnormally, even if each point alone may look normal.

When it occurs: Often in sequences or groups of events.

Example:

Multiple failed login attempts on an account within a short period → collective anomaly (possible cyberattack).

In sensor data, a sequence of slightly high readings (individually not alarming) may together indicate machine overheating.
| Type of Anomaly | Key Idea                            | Example                      |
| --------------- | ----------------------------------- | ---------------------------- |
| **Point**       | Single data point deviates          | One unusual bank transaction |
| **Contextual**  | Normal overall, abnormal in context | 30°C in winter               |
| **Collective**  | Group of points abnormal together   | Series of failed logins      |


# **Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.**

# **Answer2.**
# 1. Isolation Forest (IF)

Approach:

Based on the principle that anomalies are easier to isolate than normal points.

Builds random decision trees by recursively splitting features.

Outliers tend to be isolated in fewer splits → shorter path lengths → higher anomaly score.

Use Cases:

Large, high-dimensional datasets.

Works well when anomalies are scattered and not clustered together.

Suitable for both unlabeled tabular data and streaming anomaly detection.

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

Approach:

Groups points into clusters based on density (minimum number of points within a neighborhood).

Points in sparse regions that don’t belong to any cluster are labeled as noise (outliers).

Use Cases:

Spatial/geographical data or low-to-moderate dimensional datasets.

Detecting outliers that are far from dense clusters.

Useful when clusters are irregularly shaped.

Not great for high-dimensional data (curse of dimensionality).

3. Local Outlier Factor (LOF)

Approach:

Measures the local density deviation of a data point compared to its neighbors.

A point is considered an outlier if it has significantly lower density than its neighbors.

Use Cases:

Datasets where anomalies are local (a point is normal in global scale but abnormal relative to its neighbors).

Works well in datasets with varying density regions.

Suitable for moderate-sized datasets, not very large-scale.

| Method               | Core Idea                                   | Best For                                   | Limitations                                       |
| -------------------- | ------------------------------------------- | ------------------------------------------ | ------------------------------------------------- |
| **Isolation Forest** | Isolate anomalies via random splits         | High-dimensional, large datasets           | Less effective for clustered anomalies            |
| **DBSCAN**           | Density-based clustering, noise as outliers | Spatial / low-dim data, irregular clusters | Struggles with high-dim data, parameter sensitive |
| **LOF**              | Local density comparison                    | Local anomalies, varying densities         | Computationally heavy for big data                |


# **Question 3: What are the key components of a Time Series? Explain each with one example.**

# **Answer3:**
# Key Components of a Time Series

A time series is a sequence of data points measured at regular time intervals. The key components of a time series are:

1. Trend
- Definition: A long-term direction or pattern in the data, either increasing or decreasing.
- Example: A steady increase in the average temperature over the years due to global warming.

2. Seasonality
- Definition:  Regular fluctuations that occur at fixed intervals, often related to calendar events or seasons.
- Example: Increased sales of winter clothing in December and January every year.

3. Cyclical Component
- Definition:  Long-term cycles or waves that are not of a fixed frequency, often related to economic or business cycles.
- Example: The boom and bust cycle in the housing market, where prices rise and fall over several years.

4. Irregular or Residual Component
- Definition: Random, unpredictable events or fluctuations that are not part of the trend, seasonality, or cyclical components.
- Example: A sudden drop in stock prices due to an unexpected economic event or natural disaster.

Understanding these components is crucial for time series analysis, forecasting, and modeling. By identifying and separating these components, you can gain insights into the underlying patterns and make more accurate predictions.

# **Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?**

# **Answer4:**
# Stationarity in Time Series

A time series is said to be stationary if its statistical properties, such as the mean, variance, and autocorrelation, remain constant over time. In other words, the series exhibits no trends, seasonality, or other patterns that change over time.

Testing for Stationarity

There are several methods to test for stationarity:

1. Visual Inspection: Plotting the time series and looking for trends, seasonality, or changes in variance.
2. Summary Statistics:  Calculating summary statistics, such as the mean and variance, over different time periods.
3. Statistical Tests:
    - Augmented Dickey-Fuller (ADF) Test: A popular test for stationarity that checks for the presence of a unit root.
    - KPSS Test: A test that checks for trend stationarity.

Transforming a Non-Stationary Series

If a time series is found to be non-stationary, it can be transformed into a stationary series using various techniques:

1. Differencing: Subtracting each value from its previous value to remove trends and seasonality.
2. Log Transformation:  Taking the logarithm of the series to stabilize the variance.
3. Normalization:  Subtracting the mean and dividing by the standard deviation to normalize the series.
4. Seasonal Decomposition: Decomposing the series into trend, seasonality, and residuals, and then removing the seasonality component.
5. Differencing with Seasonality:  Differencing the series at a specific lag (e.g., 12 for monthly data with yearly seasonality) to remove seasonality.

Example

Suppose we have a time series of monthly sales data that exhibits a strong trend and seasonality. To make it stationary, we can:

1. Take the logarithm of the series to stabilize the variance.
2. Difference the series at a lag of 12 to remove seasonality.
3. Difference the series again to remove the trend.

By applying these transformations, we can convert the non-stationary series into a stationary one, which can then be modeled and forecasted using various techniques.

# **Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.**



# **Answer5:**
# Time Series Models

Here's a breakdown of the differences between AR, MA, ARIMA, SARIMA, and SARIMAX models:

1. AR (Autoregressive) Model
- Structure:  Uses past values of the time series to forecast future values.
- Equation: y_t = β_0 + β_1_y_{t-1} + … + β_p_y_{t-p} + ε_t
- Application: Suitable for time series with strong autocorrelation.

2. MA (Moving Average) Model
- Structure: Uses past errors (residuals) to forecast future values.
- Equation: y_t = μ + ε_t + θ_1_ε_{t-1} + … + θ_q_ε_{t-q}
- Application: Suitable for time series with strong moving average components.

3. ARIMA (Autoregressive Integrated Moving Average) Model
- Structure: Combines AR and MA components, with differencing to handle non-stationarity.
- Equation: ARIMA(p,d,q) = AR(p) + differencing(d) + MA(q)
- Application: Suitable for time series with trends, seasonality, and autocorrelation.

4. SARIMA (Seasonal ARIMA) Model
- Structure: Extends ARIMA by incorporating seasonal components.
- Equation: SARIMA(p,d,q)(P,D,Q)_s = ARIMA(p,d,q) + seasonal AR(P) + seasonal differencing(D) + seasonal MA(Q)
- Application: Suitable for time series with strong seasonality.

5. SARIMAX Model
- Structure: Extends SARIMA by incorporating exogenous variables (explanatory variables).
- Equation:  SARIMAX(p,d,q)(P,D,Q)_s = SARIMA(p,d,q)(P,D,Q)_s + exogenous variables
- Application: Suitable for time series with strong seasonality, trends, and exogenous variables that impact the series.

Key differences:

- AR and MA: Focus on autocorrelation and moving average components, respectively.
- ARIMA:  Combines AR and MA components, with differencing for non-stationarity.
- SARIMA:  Extends ARIMA by incorporating seasonal components.
- SARIMAX: Extends SARIMA by incorporating exogenous variables.

When choosing a model, consider the characteristics of your time series data, such as trends, seasonality, autocorrelation, and the presence of exogenous variables.

# **Question 6: Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.datasets import get_rdataset

# Load the AirPassengers dataset (monthly airline passengers 1949–1960)
data = get_rdataset("AirPassengers").data

# Convert to datetime index
data['Month'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('Month', inplace=True)

# Rename the column for clarity
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Plot original time series
plt.figure(figsize=(10,5))
plt.plot(data.index, data['Passengers'], label="AirPassengers")
plt.title("Monthly AirPassengers (1949-1960)")
plt.xlabel("Year")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()

# Decompose the time series
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative', period=12)

# Plot decomposition
decomposition.plot()
plt.show()
```

# **Question 7: Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Generate synthetic dataset (you can replace this with NYC Taxi Fare features)
np.random.seed(42)
X = 0.5 * np.random.randn(200, 2)   # Normal cluster
X = np.r_[X, np.random.uniform(low=-4, high=4, size=(20, 2))]  # Add some anomalies

# Apply Isolation Forest
clf = IsolationForest(contamination=0.1, random_state=42)
y_pred = clf.fit_predict(X)

# Convert predictions: -1 = anomaly, 1 = normal
normal = X[y_pred == 1]
anomaly = X[y_pred == -1]

# Plot results
plt.figure(figsize=(8,6))
plt.scatter(normal[:, 0], normal[:, 1], c='blue', label="Normal", alpha=0.6)
plt.scatter(anomaly[:, 0], anomaly[:, 1], c='red', label="Anomaly", marker='x', s=100)
plt.title("Isolation Forest Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

# **Question 8: Train a SARIMA model on the monthly airline passengers dataset.Forecast the next 12 months and visualize the results.**

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.datasets import get_rdataset

# Load AirPassengers dataset
data = get_rdataset("AirPassengers").data

# Convert to datetime index
data['Month'] = pd.date_range(start='1949-01', periods=len(data), freq='M')
data.set_index('Month', inplace=True)
data.rename(columns={'value': 'Passengers'}, inplace=True)

# Train SARIMA model
# Order = (p,d,q), Seasonal_order = (P,D,Q,s)
# p,d,q = ARIMA terms; P,D,Q,s = seasonal terms
model = SARIMAX(data['Passengers'],
                order=(1,1,1),
                seasonal_order=(1,1,1,12))
result = model.fit(disp=False)

# Forecast next 12 months
forecast = result.get_forecast(steps=12)
forecast_index = pd.date_range(start=data.index[-1] + pd.offsets.MonthBegin(1), periods=12, freq='M')
forecast_values = forecast.predicted_mean
conf_int = forecast.conf_int()
```

# **Question 9: Apply Local Outlier Factor (LOF) on any numerical dataset to detect anomalies and visualize them using matplotlib.**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Generate synthetic dataset (you can replace this with any numerical dataset)
np.random.seed(42)
X = 0.3 * np.random.randn(200, 2)    # Normal cluster
X = np.r_[X, np.random.uniform(low=-4, high=4, size=(20, 2))]  # Inject anomalies

# Apply Local Outlier Factor
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)

# -1 = anomaly, 1 = normal
normal = X[y_pred == 1]
anomaly = X[y_pred == -1]

# Plot results
plt.figure(figsize=(8,6))
plt.scatter(normal[:, 0], normal[:, 1], c='blue', label="Normal", alpha=0.6)
plt.scatter(anomaly[:, 0], anomaly[:, 1], c='red', marker='x', s=100, label="Anomaly")
plt.title("Local Outlier Factor (LOF) Anomaly Detection")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

# Real-Time Data Science Workflow for Power Grid Monitoring
🔹 1. Anomaly Detection in Streaming Data

* Goal: Detect sudden spikes or drops in energy usage (possible faults, theft, or unusual demand).

* Approach:

* Use unsupervised methods since anomalies are rare.

* Isolation Forest → good for high-dimensional numerical features, isolates anomalies efficiently.

* LOF (Local Outlier Factor) → detects context-based anomalies (e.g., a region’s usage unusual compared to neighbors).

* DBSCAN → clusters normal behavior; points outside clusters flagged as anomalies.

* Implementation in real-time:

* Stream data in mini-batches (e.g., 15-min windows).

* Apply anomaly detection model incrementally.

* Trigger alerts when usage deviates beyond a set threshold.

🔹 2. Short-Term Forecasting Model

* Need: Forecast next few hours of energy demand for balancing supply and avoiding blackouts.

* Best choice:

* SARIMAX (Seasonal ARIMA with exogenous variables)

* Captures trend + daily seasonality.

* Incorporates weather conditions (temperature, humidity, wind) as external regressors.

Alternative (advanced):
* LSTM / GRU neural networks for real-time deep learning forecasting (if large-scale data and GPUs are available).

🔹 3. Validation & Monitoring

* Model validation:

* Split historical data into train/test.

* Use rolling-origin evaluation (walk-forward validation) → ensures realistic forecast performance.

* Metrics: MAPE, RMSE, Precision-Recall for anomalies.

* Monitoring in production:

* Continuously track forecast error (e.g., rolling MAPE).

* Drift detection: if data distribution (weather/usage) shifts, trigger retraining.

* Anomaly detection feedback loop: flag anomalies, validate with domain experts, update thresholds.

🔹 4. Business Value & Operations Impact

* Grid Stability:
* Early anomaly detection → prevent equipment failure, overload, blackouts.

* Cost Optimization:
* Forecasting demand → optimize power generation & purchasing, reduce wastage.

* Energy Trading:
* Accurate forecasts → better decisions in buying/selling energy in real-time markets.

- Customer Insights:
* Identify abnormal consumption patterns → fraud/theft detection, billing accuracy.

- Sustainability:
* Forecast demand vs renewable supply (solar/wind) → optimize usage of clean energy.

In summary:
* Anomaly detection: Isolation Forest + LOF for streaming data.

* Forecasting: SARIMAX with weather as exogenous features.

* Validation/monitoring: Rolling evaluation, error tracking, drift detection.

* Business value: Reliable operations, reduced costs, customer trust, and better energy trading.
